##품목코드 정리

```python
#Normalization
final_df['ITEM_CD'].unique()
```

```python
final_df['ITEM_CD'] = final_df['ITEM_CD'].apply(lambda x : x.split('_')[0])
final_df['ITEM_CD'] = final_df['ITEM_CD'].apply(lambda x : x.split('-')[0])
```

```python
len(final_df['ITEM_CD'].unique())
```

```python
final_df.shape
```

```python
#수량을 제외한 컬럼은 str화
for i in ['CORP_ID', 'REF_ORD_NO', 'BKG_TYP', 'BKG_DATE', 'BKG_TIME', 'SHPR_CD',
       'ITEM_CD', 'REF_ITEM_SEQ','DLVPREARRBRANCD', 'DLVPREARREMPNICKNM', 'DLVCLSFCD',
       'DLVSUBCLSFCD', 'POST_ZONE', 'SHPR_ADDR_1', 'SHPR_ADDR_2',
       'CNEE_ADDR_1', 'CNEE_ADDR_2']:
  final_df[i] = final_df[i].apply(lambda x : str(x))
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df = final_df[['CORP_ID', 'REF_ORD_NO', 'BKG_TYP', 'BKG_DATE', 'BKG_TIME', 'SHPR_CD',
       'INV_AMT', 'ITEM_CD', 'ITEM_QTY', 'ITEM_AMT', 'REF_ITEM_SEQ',
        'DLVPREARRBRANCD', 'DLVPREARREMPNICKNM', 'DLVCLSFCD',
       'DLVSUBCLSFCD', 'POST_ZONE', 'SHPR_ADDR_1', 'SHPR_ADDR_2',
       'CNEE_ADDR_1', 'CNEE_ADDR_2']]
```

```python
#물품의 개별가격
final_df['개별가격'] = final_df['ITEM_AMT']/final_df['ITEM_QTY']
```

```python
gb = final_df.groupby('ITEM_CD').sum().reset_index()
```

```python
gb.head()
```

```python
gb = gb[['ITEM_CD','ITEM_AMT','ITEM_QTY']]
```

```python
#물품의 평균가격(날짜별 변화 반영x)
gb['평균가격'] = gb['ITEM_AMT']/gb['ITEM_QTY']
```

```python
gb = gb[['ITEM_CD','평균가격']]
```

```python
#평균가격 join
final_df = pd.merge(final_df, gb, how= 'left', on = 'ITEM_CD')
```

```python
final_df.head()
```

```python
final_df['BKG_TIME'] = final_df['BKG_TIME'].apply(lambda x: x.zfill(6))
```
